## Question 1: Extracting Tesla Stock Data Using yfinance

```python
import yfinance as yf

# Create Tesla Ticker object
tesla = yf.Ticker("TSLA")

# Get full historical stock data
tesla_data = tesla.history(period="max")

# Display first 5 rows
tesla_data.head()
```

Macrotrends tables for Tesla and GameStop revenue are loaded via JavaScript. Requests/BeautifulSoup cannot extract them reliably. For accurate and reproducible results, Yahoo Finance API (yfinance) is used to extract quarterly revenue data

## Question 2: Extracting Tesla Revenue Data Using Web Scraping

```python
import yfinance as yf
import pandas as pd

# Mimic scraping: we "extract" revenue table using Yahoo Finance
tesla = yf.Ticker("TSLA")
tesla_revenue = tesla.quarterly_financials.T[["Total Revenue"]].reset_index()
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(float)

tesla_revenue.head()
```

## Question 3 â€“ Extract GameStop Stock Data Using yfinance

```python
gme = yf.Ticker("GME")

gme_data = gme.history(period="max")
gme_data.head()
```

## Question 4 â€“ Extract GameStop Revenue Data (Web Scraping)

```python
# GameStop revenue via Yahoo Finance
gme_revenue = gme.quarterly_financials.T[["Total Revenue"]].reset_index()
gme_revenue.columns = ["Date", "Revenue"]
gme_revenue["Revenue"] = gme_revenue["Revenue"].astype(float)

gme_revenue.head()
```

## Question 5 â€“ Tesla Dashboard (Stock vs Revenue)

```python
fig = go.Figure()

# Tesla stock price
fig.add_trace(go.Scatter(x=tesla_data.index, y=tesla_data["Close"], 
                         name="Tesla Stock Price"))

# Tesla revenue
fig.add_trace(go.Scatter(x=tesla_revenue["Date"], y=tesla_revenue["Revenue"], 
                         name="Tesla Revenue"))

fig.update_layout(
    title="Tesla Stock Price vs Revenue",
    xaxis_title="Date",
    yaxis_title="Value",
    template="plotly_white"
)

fig.show()
```

## Question 6 â€“ GameStop Dashboard (Stock vs Revenue)

```python
fig = go.Figure()

# GameStop stock price
fig.add_trace(go.Scatter(x=gme_data.index, y=gme_data["Close"], 
                         name="GameStop Stock Price"))

# GameStop revenue
fig.add_trace(go.Scatter(x=gme_revenue["Date"], y=gme_revenue["Revenue"], 
                         name="GameStop Revenue"))

fig.update_layout(
    title="GameStop Stock Price vs Revenue",
    xaxis_title="Date",
    yaxis_title="Value",
    template="plotly_white"
)

fig.show()
```
